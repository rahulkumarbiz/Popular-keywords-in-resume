# Objective is to find out the most popular keywords used by Data Scientists in their resume.

First we will select top 40 resume from resume data base of Naukri.com (Resdex).Then copy all the keywords in notpad and save it as csv file.

Next step is to clean our text data which includes removing numbers and special characters.

Now we will perform EDA to bring insights and finally visualise our result in form of bar chart and word cloud.



```python
#Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#read data
df=pd.read_csv('Keywords.csv')
```

```python
col=df.columns
col
```

```python
#Converting rows to columns
df=pd.DataFrame({'keywords':col})
df
```

```python
df.info()
```

```python
df.head(10)
```

```python
#Cleaning text
import re
lst =[]
for i in range(0,867): 
    row = re.sub('[^a-zA-Z]',' ',df['keywords'][i])
    row = row.lower()
    word = row.split()  
    row = ' '.join(word)
    lst.append(row)
   
```

```python
len(lst)
```

```python
lst
```

```python
master_df=pd.DataFrame({'keywords':lst})
```

```python
#Creating clean csv file
master_df.to_csv('Final_Keywords.csv',index=False)
```

```python
x=master_df['keywords'].value_counts().head(30).keys()
```

```python
y=master_df['keywords'].value_counts().head(30).values
```

```python
#ploting most repeated 30 keyskills 
plt.figure(figsize=(15,5))
plt.style.use('ggplot')
plt.title('Counts vs Keywords')
plt.ylabel('Counts')
plt.bar(x,y,color='g')
plt.xticks(rotation=90)
plt.show()
```

```python
#These 30 skills(keywords) covers the 61.25% of Data
round(master_df['keywords'].value_counts().head(30).sum()/len(master_df.index)*100,2)
```

```python
#Creating word cloud for better visualisation
from wordcloud import WordCloud
wordcloud = WordCloud().generate_from_frequencies(master_df['keywords'].value_counts())
plt.figure(figsize = (10,10),facecolor = None )
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
